## 1.3 Combining Datasets

### 1.3.1 Combining CDR datasets

```python
# We combine the CDR datafiles
#
df_cdrs = pd.concat([df_01,df_02,df_03], axis=0)
df_cdrs.info()
print(df_cdrs)
```

### 1.3.2 Merging CDR data and Cell Geo Data

```python
#Check for duplicates
#
df_cdrs = df_cdrs.drop_duplicates(keep='first')
df_cdrs.info
```

```python
# Combine cdr data and cell geo data
#
df_cells_cdrs = pd.merge(df_cdrs, df, how='inner', left_index=True, right_index=True)
df_cells_cdrs
```
